```python
#checking unique disease
df_idsp['disease_illness_name'].unique()
```

```python
air_pollution_related = {
    'ARI Influenza Like Illness(ILI)',
    'Seasonal Influenza',
    'Swine Flu (H1N1)',
    'Acute Respiratandy Illness',
    'Influenza',
    'Influenza A',
    'H3N2',
    'Fever and Upper Respiratandy Tract Infection (URTI)',
    'Allergic Conjunctivitis',
    'Chemical Gas Poisoning',
    'HMPV',
    'Adenovirus'
}

# Classify diseases strictly based on your list
def classify_disease_strict(disease):
    if pd.isna(disease['disease_illness_name']):
        return 'Unknown'
    disease['disease_illness_name'] = disease['disease_illness_name'].strip()
    if disease['disease_illness_name'] in air_pollution_related:
        return 'Air Pollution Related'
    else:
        return 'Not Air Pollution Related'

df_idsp['pollution_relation'] = df_idsp.apply(classify_disease_strict,axis = 1)

# Check the counts
print(df_idsp['pollution_relation'].value_counts())
```

```python
df_idsp_air_pol_rel = df_idsp[df_idsp['pollution_relation']== 'Air Pollution Related']
```

```python
#no of cases based on disease caused by air pollution
df_idsp_air_pol_rel.loc[df_idsp_air_pol_rel.groupby('disease_illness_name')['cases'].idxmax().sort_values(ascending = False)][['state','cases','disease_illness_name']]
```

```python
plt.figure(figsize = (10,4))
df_idsp_air_pol_rel.groupby('disease_illness_name')['cases'].sum().sort_values(ascending = False).plot(kind = 'bar')
plt.title("No. of cases based on disease caused by air pollution")
plt.show()
```

## What are the most essential features that should be incorporated into the air purifier?

 # 1. High-Efficiency Particulate Filtration
True HEPA H13 or H14 filter to remove fine particles:

PM2.5 & PM10 from dust, traffic, and burning.

Efficiency â‰¥99.95% for particles â‰¥0.3 Î¼m.

Pre-filter for large dust particles to protect the HEPA filter.

## 2. Gas & Chemical Pollutant Removal
Since  NOâ‚‚, SOâ‚‚, Oâ‚ƒ, and CO  are also prominent pollutant:

Activated Carbon Filter for NOâ‚‚, SOâ‚‚, VOCs, and odors.

Potassium Permanganate or Chemisorbent Media to neutralize acidic gases like SOâ‚‚ and ozone (Oâ‚ƒ).

Optional Photocatalytic Oxidation (PCO) for additional VOC breakdown.

## 3. Adaptability to Seasonal Changes
Replaceable/upgradeable filter modules â€” so in high PM months (winter), heavier particulate filtration is used, and in summer, more gas filtration.

Auto-scheduler to pre-run during peak pollution hours (morning/evening traffic).
